```python
import pandas as pd

# Tạo một Series với các giá trị và nhãn (index) riêng
data = [10, 20, 30, 40]
index = ['a', 'b', 'c', 'd']
s = pd.Series(data, index=index)

print("Series 1 chiều với nhãn (index):")
print(s)

# Minh họa dạng bảng (giống 1 cột Excel) - bảng đẹp hơn, dễ nhìn hơn
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(4, 2.5))
table_data = [[idx, val] for idx, val in zip(s.index, s.values)]
col_labels = ["Index", "Value"]

# Tạo bảng với style đẹp hơn
table = ax.table(
    cellText=table_data,
    colLabels=col_labels,
    loc='center',
    cellLoc='center',
    colColours=['#4F81BD', '#4F81BD'],
    colWidths=[0.2, 0.2]
)

# Định dạng header
for (row, col), cell in table.get_celld().items():
    if row == 0:
        cell.set_fontsize(12)
        cell.set_text_props(weight='bold', color='white')
        cell.set_facecolor('#4F81BD')
    else:
        cell.set_fontsize(11)
        if row % 2 == 1:
            cell.set_facecolor('#F2F2F2')
        else:
            cell.set_facecolor('#FFFFFF')

# Định dạng border
table.auto_set_font_size(False)
table.scale(1.2, 1.2)
table.auto_set_column_width([0, 1])

# Ẩn trục và thêm tiêu đề
ax.axis('off')
plt.title("Bảng Series: 1 cột với index và value", fontsize=14, fontweight='bold', color='#333333', pad=15)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Tạo DataFrame minh họa
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Gender": ["Female", "Male", "Male", "Male"],
    "Age": [23, 30, 25, 28]
}
df = pd.DataFrame(data)

print("DataFrame (bảng 2 chiều):")
print(df)

# Minh họa bảng đẹp với nhiều cột (giống bảng Excel)
fig, ax = plt.subplots(figsize=(5, 2.5))
table_data = [[idx] + list(row) for idx, row in zip(df.index, df.values)]
col_labels = ["Index"] + list(df.columns)

# Tạo bảng với style đẹp
table = ax.table(
    cellText=table_data,
    colLabels=col_labels,
    loc='center',
    cellLoc='center',
    colColours=['#4F81BD'] * (len(col_labels)),
    colWidths=[0.13] + [0.22] * (len(col_labels)-1)
)

# Định dạng header
for (row, col), cell in table.get_celld().items():
    if row == 0:
        cell.set_fontsize(12)
        cell.set_text_props(weight='bold', color='white')
        cell.set_facecolor('#4F81BD')
    else:
        cell.set_fontsize(11)
        if row % 2 == 1:
            cell.set_facecolor('#F2F2F2')
        else:
            cell.set_facecolor('#FFFFFF')

table.auto_set_font_size(False)
table.scale(1.2, 1.2)
table.auto_set_column_width(list(range(len(col_labels))))

ax.axis('off')
plt.title("Bảng DataFrame: Nhiều cột (Name, Gender, Age) và chỉ số hàng", fontsize=14, fontweight='bold', color='#333333', pad=15)
plt.tight_layout()
plt.show()
```
